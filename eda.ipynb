```python
# Step 3: Prepare Insert Query with Placeholders
insert_query = """
INSERT INTO Cultivation_data (
    Dist_Code, Year, State_Code, State_Name, Dist_Name,
    RICE_AREA, RICE_PRODUCTION, RICE_YIELD,
    WHEAT_AREA, WHEAT_PRODUCTION, WHEAT_YIELD,
    SORGHUM_AREA, SORGHUM_PRODUCTION, SORGHUM_YIELD,
    MAIZE_AREA, MAIZE_PRODUCTION, MAIZE_YIELD,
    FINGER_MILLET_AREA, FINGER_MILLET_PRODUCTION, FINGER_MILLET_YIELD,
    GROUNDNUT_AREA, GROUNDNUT_PRODUCTION, GROUNDNUT_YIELD,
    SUNFLOWER_AREA, SUNFLOWER_PRODUCTION, SUNFLOWER_YIELD,
    SOYABEAN_AREA, SOYABEAN_PRODUCTION, SOYABEAN_YIELD,
    OILSEEDS_AREA, OILSEEDS_PRODUCTION, OILSEEDS_YIELD,
    SUGARCANE_AREA, SUGARCANE_PRODUCTION, SUGARCANE_YIELD,
    COTTON_AREA, COTTON_PRODUCTION, COTTON_YIELD,
    FRUITS_AREA, VEGETABLES_AREA, FRUITS_AND_VEGETABLES_AREA,
    POTATOES_AREA, ONION_AREA, FODDER_AREA
) VALUES (
    %s, %s, %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s, 
    %s, %s, %s
)
"""

# Iterate Over Dataset and Insert Each Row
for _, row in df.iterrows():
    values = (
        row["Dist Code"], row["Year"], row["State Code"],
        row["State Name"], row["Dist Name"],
        row["RICE AREA (1000 ha)"], row["RICE PRODUCTION (1000 tons)"], row["RICE YIELD (Kg per ha)"],
        row["WHEAT AREA (1000 ha)"], row["WHEAT PRODUCTION (1000 tons)"], row["WHEAT YIELD (Kg per ha)"],
        row["SORGHUM AREA (1000 ha)"], row["SORGHUM PRODUCTION (1000 tons)"], row["SORGHUM YIELD (Kg per ha)"],
        row["MAIZE AREA (1000 ha)"], row["MAIZE PRODUCTION (1000 tons)"], row["MAIZE YIELD (Kg per ha)"],
        row["FINGER MILLET AREA (1000 ha)"], row["FINGER MILLET PRODUCTION (1000 tons)"], row["FINGER MILLET YIELD (Kg per ha)"],
        row["GROUNDNUT AREA (1000 ha)"], row["GROUNDNUT PRODUCTION (1000 tons)"], row["GROUNDNUT YIELD (Kg per ha)"],
        row["SUNFLOWER AREA (1000 ha)"], row["SUNFLOWER PRODUCTION (1000 tons)"], row["SUNFLOWER YIELD (Kg per ha)"],
        row["SOYABEAN AREA (1000 ha)"], row["SOYABEAN PRODUCTION (1000 tons)"], row["SOYABEAN YIELD (Kg per ha)"],
        row["OILSEEDS AREA (1000 ha)"], row["OILSEEDS PRODUCTION (1000 tons)"], row["OILSEEDS YIELD (Kg per ha)"],
        row["SUGARCANE AREA (1000 ha)"], row["SUGARCANE PRODUCTION (1000 tons)"], row["SUGARCANE YIELD (Kg per ha)"],
        row["COTTON AREA (1000 ha)"], row["COTTON PRODUCTION (1000 tons)"], row["COTTON YIELD (Kg per ha)"],
        row["FRUITS AREA (1000 ha)"], row["VEGETABLES AREA (1000 ha)"], row["FRUITS AND VEGETABLES AREA (1000 ha)"],
        row["POTATOES AREA (1000 ha)"], row["ONION AREA (1000 ha)"], row["FODDER AREA (1000 ha)"]
    )
    cursor.execute(insert_query, values)

conn.commit()
print("All rows inserted successfully into 'Cultivation_data'!")
```

```python
#1.Year-wise Trend of Rice Production Across States (Top 3)
cursor.execute("""SELECT 
    Year,
    State_Name,
    Total_Rice_Production
FROM (
    SELECT 
        Year,
        State_Name,
        SUM(RICE_PRODUCTION) AS Total_Rice_Production,
        RANK() OVER (PARTITION BY Year ORDER BY SUM(RICE_PRODUCTION) DESC) AS rank_state
    FROM 
        Cultivation_data
    GROUP BY 
        Year, State_Name
) AS ranked
WHERE rank_state <= 3
ORDER BY Year , rank_state;""")
out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#2.Top 5 Districts by Wheat Yield Increase Over the Last 5 Years
cursor.execute("""SELECT 
    Dist_Name,
    (MAX(WHEAT_YIELD) - MIN(WHEAT_YIELD)) AS Yield_Increase
FROM Cultivation_data
WHERE Year >= (SELECT MAX(Year) FROM Cultivation_data) - 5
GROUP BY Dist_Name
ORDER BY Yield_Increase DESC
LIMIT 5;""")
out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#3.States with the Highest Growth in Oilseed Production (5-Year Growth Rate)
cursor.execute("""SELECT 
    State_Name,
    (MAX(OILSEEDS_PRODUCTION) - MIN(OILSEEDS_PRODUCTION)) AS Growth_5_Years
FROM Cultivation_data
WHERE Year >= (SELECT MAX(Year) FROM Cultivation_data) - 5
GROUP BY State_Name
ORDER BY Growth_5_Years DESC
LIMIT 5;""")
out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#4.District-wise Correlation Between Area and Production for Major Crops (Rice, Wheat, and Maize)
cursor.execute("""SELECT 
    Dist_Name,
    SUM(RICE_AREA) AS Total_Rice_Area,
    SUM(RICE_PRODUCTION) AS Total_Rice_Production,
    
    SUM(WHEAT_AREA) AS Total_Wheat_Area,
    SUM(WHEAT_PRODUCTION) AS Total_Wheat_Production,

    SUM(MAIZE_AREA) AS Total_Maize_Area,
    SUM(MAIZE_PRODUCTION) AS Total_Maize_Production
FROM Cultivation_data
GROUP BY Dist_Name;""")
out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#5.Yearly Production Growth of Cotton in Top 5 Cotton Producing States
cursor.execute("""SELECT 
    c.State_Name,
    c.Year,
    SUM(c.COTTON_PRODUCTION) AS Cotton_Production
FROM Cultivation_data c
JOIN (
        SELECT State_Name
        FROM Cultivation_data
        GROUP BY State_Name
        ORDER BY SUM(COTTON_PRODUCTION) DESC
        LIMIT 5
     ) AS top5
ON c.State_Name = top5.State_Name
GROUP BY c.State_Name, c.Year
ORDER BY c.State_Name, c.Year;""")
out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#6.Districts with the Highest Groundnut Production in 2020
cursor.execute("""SELECT 
    Dist_Name,
    State_Name,
    GROUNDNUT_PRODUCTION
FROM Cultivation_data
WHERE Year = 2017
ORDER BY GROUNDNUT_PRODUCTION DESC
LIMIT 10;""")
out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#7.Annual Average Maize Yield Across All States
cursor.execute("""SELECT State_Name, 
    Year, 
    Avg(MAIZE_YIELD) As  Avg_Maize_Yield
    FROM Cultivation_data 
    GROUP BY State_Name Order by Year DESC""")
out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#8.Total Area Cultivated for Oilseeds in Each State
cursor.execute("""SELECT 
    State_Name,
    SUM(OILSEEDS_AREA) AS Total_Oilseeds_Area
FROM Cultivation_data
GROUP BY State_Name
ORDER BY Total_Oilseeds_Area DESC;""")
out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#9.Districts with the Highest Rice Yield
cursor.execute("""SELECT 
    Dist_Name,
    State_Name,
    RICE_YIELD
FROM Cultivation_data
ORDER BY RICE_YIELD DESC
LIMIT 10;""")
out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#10.Compare the Production of Wheat and Rice for the Top 5 States Over 10 Years
cursor.execute("""SELECT 
    c.State_Name,
    c.Year,
    SUM(c.RICE_PRODUCTION) AS Rice_Production,
    SUM(c.WHEAT_PRODUCTION) AS Wheat_Production
FROM Cultivation_data c
JOIN (
        SELECT State_Name
        FROM Cultivation_data
        GROUP BY State_Name
        ORDER BY SUM(RICE_PRODUCTION + WHEAT_PRODUCTION) DESC
        LIMIT 5
     ) AS top5
ON c.State_Name = top5.State_Name
WHERE c.Year >= (SELECT MAX(Year) FROM Cultivation_data) - 9
GROUP BY c.State_Name, c.Year
ORDER BY c.State_Name, c.Year;""")
out = cursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers = [i[0] for i in cursor.description], tablefmt = "psql"))
```

```python
#Top 7 RICE PRODUCTION State Data(Bar_plot)

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv(file_path)

# Step 2: Group by State and calculate total rice production
top_7_rice = (
    df.groupby("State Name")["RICE PRODUCTION (1000 tons)"]
    .sum()                           # Total production per state
    .sort_values(ascending=False)    # Sort by highest production
    .head(7)                         # Top 7 states
    .reset_index()                   # Reset index for plotting
)

# Step 3: Create a Bar Plot
plt.figure(figsize=(10, 6),facecolor='#ada397')
plt.bar(top_7_rice["State Name"], top_7_rice["RICE PRODUCTION (1000 tons)"],
        color="#AEB6BF", edgecolor="#34495E")

# Step 4: Add titles and labels
plt.title("Top 7 Rice Producing States in India", fontsize=14, fontweight='bold')
plt.xlabel("State Name", fontsize=12)
plt.ylabel("Total Rice Production (1000 tons)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Step 5: Show the plot
plt.tight_layout()
plt.show()
```

```python
#Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load your dataset
df = pd.read_csv(file_path)

# Step 2: Group by state and calculate total wheat production
top_5_wheat = (
    df.groupby("State Name")["WHEAT PRODUCTION (1000 tons)"]
    .sum()                           # Sum production for each state
    .sort_values(ascending=False)    # Sort from highest to lowest
    .head(5)                         # Pick top 5 states
    .reset_index()                   # Reset index for clean plotting
)

# Step 3: Calculate each state’s percentage contribution
top_5_wheat["Percentage"] = (
    top_5_wheat["WHEAT PRODUCTION (1000 tons)"] /
    top_5_wheat["WHEAT PRODUCTION (1000 tons)"].sum()
) * 100

# ===============================
# BAR CHART — Wheat Production
# ===============================
plt.figure(figsize=(8, 6))

plt.bar(
    top_5_wheat["State Name"],
    top_5_wheat["WHEAT PRODUCTION (1000 tons)"],
    color=["#ffcc99", "#ff9966", "#ff6600", "#ff3300", "#cc0000"],
    edgecolor='black'
)

plt.title("Top 5 Wheat Producing States in India", fontsize=14, fontweight="bold")
plt.xlabel("State Name", fontsize=12)
plt.ylabel("Wheat Production (1000 tons)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# ======================================
# PIE CHART — Percentage Contribution
# ======================================
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_wheat["WHEAT PRODUCTION (1000 tons)"],
    labels=top_5_wheat["State Name"],
    autopct='%1.1f%%',
    startangle=90,
    colors=['#ffcc99', '#ff9966', '#ff6600', '#ff3300', '#cc0000']
)
plt.legend(["Uttar Pradesh","Punjab","Haryana","Madhya Pradesh","Rajasthan"],loc="upper left")
plt.title("Percentage Share of Top 5 Wheat Producing States", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
#India's Millet Production Over the Last 50 Years

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load your dataset
df = pd.read_csv(file_path)

# Step 2: Calculate total Millet production for each year
# Combining Sorghum, Pearl Millet, and Finger Millet production
df["TOTAL_MILLET_PRODUCTION"] = (
    df["SORGHUM PRODUCTION (1000 tons)"] +
    df["PEARL MILLET PRODUCTION (1000 tons)"] +
    df["FINGER MILLET PRODUCTION (1000 tons)"]
)

# Step 3: Group by year to get India's total millet production
millet_trend = (
    df.groupby("Year")["TOTAL_MILLET_PRODUCTION"]
    .sum()                      # Total millet production in India for each year
    .reset_index()
    .sort_values("Year")
)

# Step 4: Take only the last 50 years
millet_trend = millet_trend.tail(50)

# ======================================
# LINE CHART — Millet Production Over Last 50 Years
# ======================================
plt.figure(figsize=(12, 6),facecolor="#cfe0e8")

in_facecolour=plt.axes()
in_facecolour.set_facecolor("#cfe0e8")

plt.plot(
    millet_trend["Year"],
    millet_trend["TOTAL_MILLET_PRODUCTION"],
    color="saddlebrown",
    marker="D",
    linewidth=0.5,
    markersize=7
)

# Step 5: Add titles and labels
plt.title("India's Millet Production Over the Last 50 Years", fontsize=14, fontweight="bold")
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Millet Production (1000 tons)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Sorghum Production (Kharif and Rabi) by Region

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load your dataset
df = pd.read_csv(file_path)

# Step 2: Group by State and calculate total Sorghum production for Kharif & Rabi
sorghum_region = (
    df.groupby("State Name")[["KHARIF SORGHUM PRODUCTION (1000 tons)", 
                              "RABI SORGHUM PRODUCTION (1000 tons)"]]
    .sum()
    .reset_index()
)

# ======================================
# Stacked Bar Plot
# ======================================

plt.figure(figsize=(12, 6))

# Plot Kharif first
plt.bar(
    sorghum_region["State Name"],
    sorghum_region["KHARIF SORGHUM PRODUCTION (1000 tons)"],
    label="Kharif",
    color="#96ceb4"    
)

# Plot Rabi on top (stacked)
plt.bar(
    sorghum_region["State Name"],
    sorghum_region["RABI SORGHUM PRODUCTION (1000 tons)"],
    bottom=sorghum_region["KHARIF SORGHUM PRODUCTION (1000 tons)"],
    label="Rabi",
    color="#5b9aa0"     
)

# Customize chart appearance
plt.title("Sorghum Production (Kharif vs Rabi) by State (Stacked Bar Chart)", fontsize=14, fontweight="bold")
plt.xlabel("State Name", fontsize=12)
plt.ylabel("Production (1000 tons)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title="Season", fontsize=10)
plt.tight_layout()

plt.show()
```

```python
# Top 7 States for Groundnut Production — Donut Chart

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv(file_path)

# Step 2: Group by state and calculate total groundnut production
top_groundnut_states = (
    df.groupby("State Name")["GROUNDNUT PRODUCTION (1000 tons)"]
    .sum()
    .sort_values(ascending=False)
    .head(7)
    .reset_index()
)

# Color palette for 7 states
colors = ['#ffcc99', '#ffb366', '#ff944d', '#ff751a', '#e65c00', '#cc5200', '#b34700']

# Step 3: Create Donut Chart
plt.figure(figsize=(8, 8))

wedges, texts, autotexts = plt.pie(
    top_groundnut_states["GROUNDNUT PRODUCTION (1000 tons)"],
    labels=top_groundnut_states["State Name"],
    autopct='%1.1f%%',
    startangle=90,
    pctdistance=0.85,
    colors=colors
)

# Center circle for donut shape
centre_circle = plt.Circle((0, 0), 0.60, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Title
plt.title("Top 7 States for Groundnut Production in India (Donut Chart)", fontsize=14, fontweight="bold")
plt.tight_layout()

plt.show()
```

```python
# Simple Soybean Production and Yield Chart

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv(file_path)

# Step 2: Group by state and get top 5 soybean producing states
soybean_stats = (
    df.groupby("State Name")[["SOYABEAN PRODUCTION (1000 tons)", "SOYABEAN YIELD (Kg per ha)"]]
    .agg({
        "SOYABEAN PRODUCTION (1000 tons)": "sum",
        "SOYABEAN YIELD (Kg per ha)": "mean"
    })
    .sort_values(by="SOYABEAN PRODUCTION (1000 tons)", ascending=False)
    .head(5)
    .reset_index()
)

# Step 3: Create figure
fig, ax1 = plt.subplots(figsize=(10, 5))

# Bar chart for production
ax1.bar(
    soybean_stats["State Name"],
    soybean_stats["SOYABEAN PRODUCTION (1000 tons)"],
    color="goldenrod"
)
ax1.set_xlabel("State Name")
ax1.set_ylabel("Soybean Production (1000 tons)", color="goldenrod")
ax1.tick_params(axis='y', labelcolor='goldenrod')

# Line chart for yield efficiency
ax2 = ax1.twinx()
ax2.plot(
    soybean_stats["State Name"],
    soybean_stats["SOYABEAN YIELD (Kg per ha)"],
    color="green",
    marker="o"
)
ax2.set_ylabel("Yield Efficiency (Kg per ha)", color="green")
ax2.tick_params(axis='y', labelcolor='green')

plt.title("Soybean Production and Yield Efficiency (Top 5 States)")
plt.tight_layout()
plt.show()
```

```python
#Oilseed Production in Major States

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv(file_path)

# Step 2: Group by State and calculate total oilseed production
oilseed_states = (
    df.groupby("State Name")["OILSEEDS PRODUCTION (1000 tons)"]
    .sum()                           # Total oilseed production per state
    .sort_values(ascending=False)    # Sort from highest to lowest
    .head(10)                        # Top 10 major states
    .reset_index()
)

# ======================================
# Step 3: EDA Visualization (Pie Chart)
# ======================================
plt.figure(figsize=(8, 8))
colors = plt.cm.YlOrBr(range(50, 250, 20))  # Yellow-Orange-Brown color map

plt.pie(
    oilseed_states["OILSEEDS PRODUCTION (1000 tons)"],
    labels=oilseed_states["State Name"],
    autopct="%1.1f%%",
    startangle=90,
    colors=colors,
    shadow=True,
)

plt.title("Oilseed Production Share by Major States in India", fontsize=14, fontweight="bold", pad=15)
plt.tight_layout()
plt.show()
```

```python
#Impact of Area Cultivated on Production (Rice, Wheat, Maize)

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv(file_path)

# Step 2: Group by Year and sum total Area & Production for each crop
trend_data = (
    df.groupby("Year")[[
        "RICE AREA (1000 ha)", "RICE PRODUCTION (1000 tons)",
        "WHEAT AREA (1000 ha)", "WHEAT PRODUCTION (1000 tons)",
        "MAIZE AREA (1000 ha)", "MAIZE PRODUCTION (1000 tons)"
    ]]
    .sum()
    .reset_index()
    .sort_values("Year")
)

# Step 3: Keep only the last 50 years (for a clearer modern trend)
trend_data = trend_data.tail(50)

# ======================================
# Step 4: Line Plot for Area vs Production
# ======================================
plt.figure(figsize=(12, 7))
plt.style.use('seaborn-v0_8-darkgrid')

# --- Rice ---
plt.plot(
    trend_data["Year"], trend_data["RICE AREA (1000 ha)"], 
    color='seagreen', linestyle='--', linewidth=2, label="Rice Area"
)
plt.plot(
    trend_data["Year"], trend_data["RICE PRODUCTION (1000 tons)"], 
    color='darkgreen', linestyle='-', linewidth=2, label="Rice Production"
)

# --- Wheat ---
plt.plot(
    trend_data["Year"], trend_data["WHEAT AREA (1000 ha)"], 
    color='goldenrod', linestyle='--', linewidth=2, label="Wheat Area"
)
plt.plot(
    trend_data["Year"], trend_data["WHEAT PRODUCTION (1000 tons)"], 
    color='orange', linestyle='-', linewidth=2, label="Wheat Production"
)

# --- Maize ---
plt.plot(
    trend_data["Year"], trend_data["MAIZE AREA (1000 ha)"], 
    color='royalblue', linestyle='--', linewidth=2, label="Maize Area"
)
plt.plot(
    trend_data["Year"], trend_data["MAIZE PRODUCTION (1000 tons)"], 
    color='navy', linestyle='-', linewidth=2, label="Maize Production"
)

# ======================================
# Step 5: Chart Customization
# ======================================
plt.title("Impact of Area Cultivated on Production (Rice, Wheat, Maize) — Last 50 Years", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Area / Production (1000 ha / 1000 tons)", fontsize=12)
plt.legend(fontsize=10, loc='upper left', ncol=2)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#Rice vs. Wheat Yield Across States

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv(file_path)

# Step 2: Group by State and calculate average yield (Kg per ha)
yield_states = (
    df.groupby("State Name")[["RICE YIELD (Kg per ha)", "WHEAT YIELD (Kg per ha)"]]
    .mean()  # Average yield per state
    .sort_values(by="RICE YIELD (Kg per ha)", ascending=False)
    .head(10)  # Top 10 states for better visualization
    .reset_index()
)

# ======================================
# Step 3: Create a Stacked Bar Chart
# ======================================
plt.figure(figsize=(12, 7))
plt.style.use('seaborn-v0_8-whitegrid')

# Bar widths and stacked structure
plt.bar(
    yield_states["State Name"],
    yield_states["RICE YIELD (Kg per ha)"],
    color="mediumseagreen",
    edgecolor="black", 
    label="Rice Yield"
)

plt.bar(
    yield_states["State Name"],
    yield_states["WHEAT YIELD (Kg per ha)"],
    bottom=yield_states["RICE YIELD (Kg per ha)"],  # Stack on top of rice
    color="goldenrod",
    edgecolor="black",
    label="Wheat Yield"
)

# ======================================
# Step 4: Customize Chart
# ======================================
plt.title("Rice vs Wheat Yield Across Indian States", fontsize=15, fontweight="bold", pad=15)
plt.xlabel("State Name", fontsize=12)
plt.ylabel("Average Yield (Kg per ha)", fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.legend(title="Crop Type", fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Add yield labels on bars
for i, (rice_yield, wheat_yield) in enumerate(zip(yield_states["RICE YIELD (Kg per ha)"], yield_states["WHEAT YIELD (Kg per ha)"])):
    total = rice_yield + wheat_yield
    plt.text(i, total + 50, f"{total:,.0f}", ha='center', fontsize=9, color='black', fontweight='bold')

plt.tight_layout()
plt.show()
```
